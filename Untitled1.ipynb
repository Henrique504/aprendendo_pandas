```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('varejo.xlsx')
```

```python
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
```

```python
vendas['estado'].fillna('MS', inplace=True)
```

```python
vendas.info()
```

```python
mediaPreço = vendas['Preço'].mean()
```

```python
vendas['Preço'].fillna(mediaPreço, inplace=True)
```

```python
vendas[['Preço', 'Preço_com_frete']].describe()
```

```python
preço_errado = vendas.query('Preço > Preço_com_frete')
preço_correto = vendas.query('Preço < Preço_com_frete')
```

```python
preço_correto.describe()
```

```python
preço_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
preço_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(preço_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
preço_correto.groupby('Data').idcompra.nunique().sort_values(ascending=False)
```

```python
preço_correto['mes'] = preço_correto['Data'].dt.month
preço_correto.head()
```

```python
preço_correto.groupby('mes').idcompra.nunique().sort_values(ascending=False)
```

```python
cliente = pd.read_excel('cliente_varejo.xlsx')
```

```python
cliente = cliente.astype({'renda':'float'})
```

```python
vendas_cliente = preço_correto.merge(cliente, how='left', on='cliente_Log')
vendas_cliente.head()
```

```python
agg_idcanal_renda= round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(8,4))
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color='blue')
plt.ylabel('Média de idade')
plt.title('Idade Média por bandeira', loc='center', fontweight='bold', color='purple')
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'])
plt.ylabel('Renda média por canal de venda')
plt.title('Renda média por canal de venda')
```

```python
vendas_por_data = preço_correto.groupby('Data').idcompra.nunique().reset_index()
vendas_por_data
```

```python
fig = plt.figure(figsize=(15,10))
plt.xlabel("Data da venda")
plt.ylabel("Quantidade de vendas")
x = vendas_por_data['Data']
y = vendas_por_data['idcompra']
plt.plot(x,y, color='blue')
plt.show()
```

```python
agg_dpt_preço = round(preço_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_dpt_preço
```

```python
fig = plt.figure(figsize=(30,20))
plt.bar(agg_dpt_preço['Nome_Departamento'], agg_dpt_preço['Preço_com_frete'])
plt.ylabel('Preço médio')
plt.title('Preço médio por departamento')
```
